```python
def fft_sample(sig, fs):
    N = len(sig)  # Number of samples
    t = np.arange(N) / fs  # Time vector    
    # Compute the FFT
    fft_result = np.fft.fft(sig)
    fft_magnitude = np.abs(fft_result) / N  # Normalize
    fft_phase = np.arctan(fft_result.imag/fft_result.real)
    freq_bins = np.fft.fftfreq(N, d=1/fs)
    
    positive_freqs = freq_bins[:N//2]
    positive_magnitude = fft_magnitude[:N//2]

    return freq_bins, fft_magnitude, fft_phase

```

```python
sample_size=1000 #size of the random window we are taking 
X = []
y = []
for i in range(10):
    for e in bench: #e is the set
        t = []
        start_idx = np.random.randint(0, len(e[features[0]]) - sample_size + 1)
        for a in features:
            t.append(e[a][start_idx:start_idx+sample_size].to_numpy())
        X.append(np.nan_to_num(t[0]-t[1]))
        y.append('bench')
#X = np.array(X)

#X.shape
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

mlp_classifier = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=500, random_state=42)

mlp_classifier.fit(X_train, y_train)

y_pred = mlp_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

```
