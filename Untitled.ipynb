```python
def fft_sample(sig, fs):
    N = len(sig)  # Number of samples
    t = np.arange(N) / fs  # Time vector    
    # Compute the FFT
    fft_result = np.fft.fft(sig)
    fft_magnitude = np.abs(fft_result) / N  # Normalize
    fft_phase = np.arctan(fft_result.imag/fft_result.real)
    freq_bins = np.fft.fftfreq(N, d=1/fs)
    
    positive_freqs = freq_bins[:N//2]
    positive_magnitude = fft_magnitude[:N//2]

    return freq_bins, fft_magnitude, fft_phase

```
